```
#Crescimento população de 1980-2016
#DataSus

#visualição dos dados
import matplotlib.pyplot as plt


#abre o arquivo e faz a leitura de linhas
dados = open("original.csv").readlines()

#vetor para guardar os anos
x = []

#vetor para guardar a quantidade de pessoas
y = []


for i in range(len(dados)):
  if i!=0:
      linha = dados[i].split(";")
      x.append(int(linha[0]))
      y.append(int(linha[1]))


titulo = "Crescimento Polulacional(1980-2016)"
eixox = "anos"
eixoy = "pessoas x100.000.000"



plt.title(titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.plot(x,y, linestyle="--")
plt.bar(x,y,color="gray")
plt.show()
```
